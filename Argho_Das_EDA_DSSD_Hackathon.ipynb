Tokenized the dataset in 1 word (Frequency)

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import string
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('stopwords')

excel_file_path = '/content/Data.xlsx'
df = pd.read_excel(excel_file_path)

text_column_name = 'Review'

text_data = ' '.join(df[text_column_name].astype(str))

translator = str.maketrans('', '', string.punctuation)
text_data_no_punct = text_data.translate(translator)
tokens = word_tokenize(text_data_no_punct)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

fdist = FreqDist(filtered_tokens)

fdist.plot(20, cumulative=False)
plt.show()
```

Tokenized the dataset [All Lowercase] in 5 word (Frequency)

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import string
from nltk import ngrams
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('stopwords')

excel_file_path = '/content/Data.xlsx'
df = pd.read_excel(excel_file_path)

df['Review'] = df['Review'].astype(str).str.lower()

text_data = ' '.join(df['Review'])

translator = str.maketrans('', '', string.punctuation)
text_data_no_punct = text_data.translate(translator)

# Tokenize into trigrams
tokens = ngrams(word_tokenize(text_data_no_punct), 5)

# Convert trigrams to a list of strings
trigrams = [' '.join(token) for token in tokens]

# Remove stopwords from trigrams
stop_words = set(stopwords.words('english'))
filtered_trigrams = [trigram for trigram in trigrams if all(word.lower() not in stop_words for word in trigram.split())]

fdist = FreqDist(filtered_trigrams)

fdist.plot(20, cumulative=False)
plt.show()
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd

# Download the VADER lexicon if not already present
nltk.download('vader_lexicon')

def analyze_sentiment(sentence):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(sentence)

    if sentiment_scores['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

output_file_path = '/content/Data.xlsx'
df = pd.read_excel(output_file_path)

column_to_analyze = 'Review'
df['Sentiment'] = df[column_to_analyze].apply(analyze_sentiment)

output_file_path = 'output_with_sentiment.xlsx'
df.to_excel(output_file_path, index=False)

print(f'Sentiment analysis completed. Results saved to {output_file_path}')
```
